```python
import scipy.io
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def read_mat_file(file_name):
    try:
        data = scipy.io.loadmat(file_name)
        if 'data' in data:
            return data['data']
        else:
            print(f"Error: No 'data' key found in {file_name}.")
    except FileNotFoundError:
        print(f"Error: File {file_name} not found.")
    return None

def find_transition_point_simple(timeseries, threshold=0.9):
    for i, value in enumerate(timeseries):
        if value > threshold:
            return i
    return None

def plot_time_series(timeseries, transition_point, file_name):
    plt.figure(figsize=(10, 5))
    plt.plot(timeseries, label='Time Series')
    if transition_point is not None:
        plt.axvline(x=transition_point, color='r', linestyle='--', label=f'Transition Point: {transition_point}')
    plt.title(f'Time Series Plot for {file_name}')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

mat_file_names = [
    'x003512.mat', 'x003514.mat', 'x003660.mat',
    'x003680.mat', 'x003683.mat', 'x003696.mat',
    'x003706.mat', 'x003724.mat', 'x003750.mat',
    'x003761.mat', 'x003805.mat', 'x003821.mat',
    'x003857.mat', 'x003862.mat', 'x003866.mat',
    'x003884.mat', 'x003894.mat', 'x001004.mat',
]

for file_name in mat_file_names:
    timeseries = read_mat_file(file_name)
    if timeseries is not None:
        timeseries = timeseries.flatten()
        transition_point = find_transition_point_simple(timeseries)

        if transition_point is not None:
            print(f"File {file_name} transition point: {transition_point}")
        else:
            print(f"File {file_name} has no transition point found.")

        plot_time_series(timeseries, transition_point, file_name)
```
